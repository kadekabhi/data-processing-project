mengimpor kelas TfidfVectorizer dari library Scikit-learn (sklearn), yang digunakan untuk mengubah koleksi dokumen teks menjadi matriks fitur TF-IDF (Term Frequency-Inverse Document Frequency). TF-IDF adalah teknik yang umum digunakan dalam pemrosesan bahasa alami dan pembelajaran mesin untuk menilai pentingnya kata-kata dalam dokumen relatif terhadap kumpulan dokumen. Kelas ini membantu dalam mengekstraksi fitur dari teks yang dapat digunakan untuk berbagai tugas seperti klasifikasi teks, pengelompokan, dan pencarian informasi.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
```

Siapkan Dokumen (corpus)

```python
documents = [
    "Saya suka belajar machine learning",        # Dokumen 0
    "Belajar data science itu seru",             # Dokumen 1
    "Machine learning adalah cabang dari AI",    # Dokumen 2
    "Saya suka data data data"                   # Dokumen 3
]
```

Buat dan Latih Model TF-IDF
Sekarang kita akan membuat objek TfidfVectorizer. Objek inilah yang akan melakukan semua perhitungan TF-IDF untuk kita. Setelah itu, kita akan "melatih" (fit) dan "mengubah" (transform) dokumen kita menjadi matriks TF-IDF


```python
tfidf_vectorizer = TfidfVectorizer()

tfidf_matrix = tfidf_vectorizer.fit_transform(documents)
```

tfidf_matrix sekarang berisi skor TF-IDF untuk setiap kata di setiap dokumen. Mari kita tampilkan hasilnya dalam format yang mudah dibaca menggunakan library Pandas.

```python
feature_names = tfidf_vectorizer.get_feature_names_out()

df_tfidf = pd.DataFrame(tfidf_matrix.toarray(), columns=feature_names)

print("Dokumen Asli")
for i, doc in enumerate(documents):
    print(f"Dokumen {i}: {doc}")

print("\nMatriks TF-IDF:")
print(df_tfidf)
```

```python
top_n = 5  # mau ambil berapa kata terpenting

for i in range(len(documents)):
    print(f"\nDokumen {i}: {documents[i]}")
    
    row = df_tfidf.iloc[i]
    top_features = row.sort_values(ascending=False).head(top_n)
    
    print("Top kata berdasarkan TF-IDF:")
    for word, score in top_features.items():
        if score > 0:
            print(f"{word} : {score:.4f}")
```
